```
#### creating fg masks ####
folder_images = "foreground_only_scaled_png_200/"
target_folder = "foreground_only_scaled_png_masks/"
size_images = dict()
for dirpath, _, filenames in os.walk(folder_images):
    for path_image in filenames:
        image = os.path.abspath(os.path.join(dirpath, path_image))
        img = Image.open(image)
        bl_im = Image.new('RGB', (60,60), (0, 0, 0))
        bl_im.paste(img)
        pixels = bl_im.load() # create the pixel map
        for i in range(bl_im.size[0]): # for every pixel:
            for j in range(bl_im.size[1]):
                if pixels[i,j] != (0,0,0): # if not black:
                    pixels[i,j] = (255, 255, 255) # change to white
        bl_im.save(target_folder + path_image.split('.')[0] + '_mask.jpg')
```

```
#### create overlaid images and there masks ####
import random
bg_folder = "bg_scld_opmzd/"
fg_folder = "foreground_only_scaled_png_200/"
output_folder = "output_folder/"
overlaid_mask_folder = "overlaid_mask_folder/"

i = 0
for bg_dirpath, _, bg_filenames in os.walk(bg_folder):
    for fg_dirpath, _, fg_filenames in os.walk(fg_folder):
        for bg_image_name in bg_filenames:
            for fg_image_name in fg_filenames:
                bg_image_path = os.path.abspath(os.path.join(bg_dirpath, bg_image_name))
                fg_image_path = os.path.abspath(os.path.join(fg_dirpath, fg_image_name))
                for j in range(20):
                    b_img = Image.open(bg_image_path)
                    f_img = Image.open(fg_image_path)
                    black_img = Image.new('RGB', (224,224), (0, 0, 0))
                    x_offset=random.randint(5, 165)
                    y_offset=random.randint(5, 165)
                    black_img.paste(f_img,(x_offset,y_offset),f_img)
                    b_img.paste(f_img,(x_offset,y_offset),f_img)
                    pixels = black_img.load()
                    for p in range(black_img.size[0]): # for every pixel:
                        for q in range(black_img.size[1]):
                            if pixels[p,q] != (0,0,0): # if not black:
                                pixels[p,q] = (255, 255, 255) # change to white
                    b_img.save(output_folder + bg_image_name.split('.')[0] + '_' + fg_image_name.split('.')[0] + '_' + str(j) +'.jpg')
                    black_img.save(overlaid_mask_folder + bg_image_name.split('.')[0] + '_' + fg_image_name.split('.')[0] + '_' + str(j) +'.jpg')
                    print(i)
                    i+=1
                   
```

```
zipname = 'bg_fg_images'

def zipfolder(foldername, target_dir):            
    zipobj = zipfile.ZipFile(foldername + '.zip', 'w', zipfile.ZIP_DEFLATED)
    rootlen = len(target_dir) + 1
    for base, dirs, files in os.walk(target_dir):
        for file in files:
            fn = os.path.join(base, file)
            zipobj.write(fn, fn[rootlen:])

zipfolder(zipname, 'output_folder/')
```

```
file_name = "bg_fg_images.zip"

from googleapiclient.http import MediaFileUpload
from googleapiclient.discovery import build

auth.authenticate_user()
drive_service = build('drive', 'v3')

def save_file_to_drive(name, path):
  file_metadata = {'name': name, 'mimeType': 'application/octet-stream'}
  media = MediaFileUpload(path, mimetype='application/octet-stream', resumable=True)
  created = drive_service.files().create(body=file_metadata, media_body=media, fields='id').execute()
  
  return created

save_file_to_drive(file_name, file_name)
```
